## 1. Importación de los datos y del modelo

Este notebook contiene todo el proceso de predicción del dataset de competición una vez creado el modelo final e importado del anterior Jupyter Notebook.

Primeramente, se importa el [dataset de competición](attrition_competition_02.csv.gz) mencionado en el primer Notebook para visualizar su contenido y modificarlo.

```python
# Importamos librerías
import joblib
import pandas as pd

# Cargamos el modelo y los datos
modelo_final = joblib.load("model.pkl")
nuevos_datos = pd.read_csv("data/attrition_competition_02.csv.gz", compression="gzip", sep=",")

# Extraemos las variables que no sirven de nada
nuevos_datos.drop(columns="EmployeeID")
```

## 2. Predicciones de los datos de competición

A continuación, se procede a la __predicción del conjunto de competición__:

```python
# Realizamos las predicciones
predicciones = modelo_final.predict(nuevos_datos)

# Guardamos las predicciones en un archivo CSV, convirtiendo los valores 0 y 1 a "No" y "Yes"
df_predicciones = pd.DataFrame(predicciones, columns=["Attrition"])
df_predicciones = df_predicciones.replace({0: "No", 1: "Yes"})
df_predicciones.to_csv("data/predictions.csv", index=False)
```

Finalmente, se comprueban el número de instancias clasificadas para cada clase:
